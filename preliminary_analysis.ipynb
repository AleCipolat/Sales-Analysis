**Stratisfied sampling:**

Aplicar un muestreo estratificado basado en las diferentes categorias.
* item category
* market category
* Shop name
* item_cnt_day

```python
dict_mod_df["sales_df"]["year"].value_counts()
```

```python
# Building a training set and a test set based on a stratified sampling
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(dict_mod_df["sales_df"], dict_mod_df["sales_df"]["market_category"]):
    strat_train_set = dict_mod_df["sales_df"].loc[train_index]
    strat_test_set = dict_mod_df["sales_df"].loc[test_index]
```

```python
# put the dataframes into a single data structure
dict_sales_df = {"strat_train_set":strat_train_set, 
                 "strat_test_set":strat_test_set,
                 "sales_df":dict_mod_df["sales_df"]}
```

Validation of the sampling data versus full data

```python
# overall vs stratified sampling
overall = (dict_sales_df["sales_df"]["market_category"].value_counts() / len(dict_sales_df["sales_df"]["market_category"]) *100).to_frame() 
stratified = (dict_sales_df["strat_train_set"]["market_category"].value_counts() / len(dict_sales_df["strat_train_set"]["market_category"]) *100).to_frame()


ov_vs_strat = pd.concat({"overall":overall, "stratified":stratified},axis=1)
ov_vs_strat
```
